```python
import os
import yaml
from tranql_jupyter import KnowledgeGraph
from functools import reduce
from build_complex import get_dataset, make_model
```

### Load the response set
The response set being used includes various queries dealing with diseases that affect the airways of the lungs. We'll print out exactly what these queries are after loading the set.

Then, we union all of the individual responses into a single knowledge graph that can be turned into a dataset.

```python
# Executing in notebooks directory
with open(os.path.join("../", "data", "response_sets", "airway_lung_disease.yaml"), "r") as file:
    # Load a response set, which is a list of responses to TranQL queries.
    response_set = yaml.safe_load(file)

knowledge_graphs = []
for resp in response_set:
    print(resp["query"])
    knowledge_graphs.append(KnowledgeGraph(resp["response"]["knowledge_graph"]))
union_graph = reduce(lambda a, b: a+b, knowledge_graphs)
```

### Looking at the dataset
Let's have a quick look at what all the graphs look like when melded together. This is gonna be the dataset we train the model on, and we can make some observations about it beforehand.

```python
union_graph.render_force_graph_3d()
```

**The most important feature to note about the graph is that its components seem very well-connected.** A bunch of completely disconnected graphs (share no nodes/edges) fed as a singular dataset into the model is no better than simply training 4 separate models on each individual graph. The fact that the constituent graphs seem to share a considerable amount of nodes and edges with each other suggests that the model should be able to learn richer relationships overall.

**Before we proceed, let's also check to see *just* how many nodes and edges these graphs have in common.**

```python
node_total = 0
edge_total = 0
for k_graph in knowledge_graphs:
    node_total += len(k_graph.net.nodes)
    edge_total += len(k_graph.net.edges)
    
union_nodes = len(union_graph.net.nodes)
union_edges = len(union_graph.net.edges)

shared_nodes = node_total-union_nodes
shared_edges = edge_total-union_edges

per_shared_nodes = str(round((shared_nodes/union_nodes)*100, 1)) + "%"
per_shared_edges = str(round((shared_edges/union_edges)*100, 1)) + "%"

print("Total nodes:", node_total, "\n", "Total edges:", edge_total, "\n")
print("Union nodes:", union_nodes, "\n", "Union edges:", union_edges, "\n")
print("Shared nodes:", shared_nodes, "\n", "Shared edges:", shared_edges, "\n")
print("Percent of nodes shared:", per_shared_nodes, "\n", "Percent of edges shared:", per_shared_edges, "\n")
```

Impressive. More than 1 in 4 nodes are shared between at least two of the graphs. This is only the case for 5% of edges, but that's understandable.

**Final thing: let's also take a look at if any nodes appear in a lot of the graphs. This response set is only made up of 3 queries, so we'll define a lot as at least two of the three.** This would indicate that they're probably pretty important. It can also help give us a clue of where to look in the graph if interesting testing its predictions.

```python
prominent_nodes = []
graph_min = 2
for node in union_graph.net.nodes:
    appears_in = 0
    for k_graph in knowledge_graphs:
        if k_graph.get_node_by_id(node) != None:
            appears_in += 1
    
    if appears_in >= graph_min:
        prominent_nodes.append(union_graph.get_node_by_id(node))

# Only print the first 5 to avoid flooding the notebook
print_cutoff = 5

print(len(prominent_nodes), "nodes:")
for node in prominent_nodes[:print_cutoff]: print(node["name"])
if len(prominent_nodes) > 5: print("...")
    
# Let's also see what type the nodes are
KnowledgeGraph.plot_node_type_distributions(
    (KnowledgeGraph({"nodes": prominent_nodes, "edges": []}), "Prominent nodes")
)
```

So, about 1/6 of all nodes appear in at least 2 of the 3 graphs in the response set. It's also helpful to know that a lot of these are chemical substances, but there's also some genes in the mix as well.
